# Documentación del proyecto MLOps: Heart Disease

Este notebook resume las etapas desarrolladas en el proyecto de predicción de enfermedad cardíaca. El objetivo fue construir un flujo local de Machine Learning con buenas prácticas de validación, despliegue, pruebas y monitoreo.

## 1. Dataset y objetivo

Se utilizó el archivo `heart.csv`, que contiene variables clínicas de pacientes. La variable objetivo es `HeartDisease`, donde `1` representa presencia de enfermedad cardíaca y `0` ausencia de enfermedad cardíaca.

El objetivo del modelo fue estimar la probabilidad de enfermedad cardíaca a partir de las variables disponibles en el dataset. Para esto primero se realizo el Análisis exploratorio de datos del dataset de `heart.csv`.

## 2. Prevención de data leakage

Primero se realizó una demostración de fuga de datos mediante una variable artificial llamada `leaky_feature`, construida a partir de la variable objetivo. Este caso mostró que el modelo puede obtener métricas artificialmente altas cuando recibe información que no estaría disponible en una predicción real.

Luego se aplicó un flujo correcto, donde la división entre entrenamiento y prueba se hizo antes del preprocesamiento. Las transformaciones se incluyeron dentro de un `Pipeline`, evitando que el conjunto de prueba aportara información al entrenamiento.

## 3. Modelado seguro

El entrenamiento se realizó usando `Pipeline` y `GridSearchCV`. Esto permitió ajustar el preprocesamiento y los modelos dentro de la validación cruzada.

Se compararon varios clasificadores, entre ellos:

- SVC
- Logistic Regression
- Ridge Logistic Regression
- Lasso Logistic Regression
- KNN
- Random Forest
- Gradient Boosting
- XGBoost

La comparación se hizo con AUC y Accuracy. El mejor modelo fue seleccionado con base en el ranking de resultados y luego se evaluó con matriz de confusión, curva ROC y AUC final.

## 4. FastAPI y Docker

El mejor modelo se guardó como `model.joblib` dentro de la carpeta `app/`. Luego se creó una API con FastAPI para recibir los datos clínicos de un paciente y devolver:

- Probabilidad estimada de enfermedad cardíaca.
- Clase predicha (`0` o `1`).

La API se probó localmente en:

```text
http://127.0.0.1:8000/docs
```

Después se creó una imagen Docker con el archivo `docker/Dockerfile`. Esto permitió ejecutar la API dentro de un contenedor, haciendo el proyecto más reproducible.

## 5. Kubernetes local con Minikube

Después de probar Docker, el contenedor fue desplegado localmente con Kubernetes usando Minikube.

Se crearon dos archivos:

- `k8s/deployment.yaml`: indica cómo ejecutar la API.
- `k8s/service.yaml`: expone la API para poder acceder desde el navegador.

El pod quedó en estado `Running`, lo que indica que la aplicación fue desplegada correctamente en Kubernetes local.

## 6. Integración continua

Se agregó un workflow de GitHub Actions en:

```text
.github/workflows/ci.yml
```

Este workflow instala las dependencias, ejecuta `flake8` para revisar estilo y corre las pruebas con `pytest`.

También se crearon pruebas básicas para verificar que la API responda correctamente en la ruta principal `/` y en el endpoint `/predict`. Localmente, las pruebas pasaron correctamente.

## 7. Monitoreo con Evidently

Se generó un reporte de deriva de datos con Evidently, comparando:

- `X_train` como datos de referencia.
- `X_test` como datos actuales.

El reporte se guardó como:

```text
drift_report.html
```

Este archivo permite revisar si existen diferencias importantes entre las distribuciones de entrenamiento y prueba.

## 8. Conclusión

El proyecto integró varias etapas de un flujo MLOps local: prevención de fuga de datos, entrenamiento seguro, comparación de modelos, despliegue con API, contenerización, orquestación local, pruebas automáticas y monitoreo de deriva.
